```python
import pandas as pd
import numpy as np
import random

   


repititions = 100000
size = 101
step = 10
data = []
results = []
for i in range(size):
    results.append([])
    for j in range(size):
        data.append([i, j, 0, 0, 0])
        results[i].append(0)
        
print('results:',results)
        
#print(random.random())
#print(data)

def fight(dead, living):
    p = random.random()
    if p <= 0.5:
        #print('fightdead: ', dead)
        dead -= 1
    else:
        #print('fightliving: ',living)
        living -= 1
        dead += 1 
    return [dead, living]

    
def store_result(loc, dead_win):
    if dead_win == True:
       # print('dead won!')
        data[loc][2] += 1
    else:
      #  print('living_won!')
        data[loc][3] += 1
        
def battle(i):
    dead = 0
    living = 0
    for j in range(data[i][0]):
        dead += 1
    for j in range(data[i][1]):
        living += 1
    
    #print('dead: ', len(dead), ', living: ',len(living))
    #print(living)
    while (dead > 0 and living > 0):
      #  print('whiledead: ',dead,'whileliving: ' ,living)
        rf = fight(dead, living)
        dead = rf[0]
        living = rf[1]
    else:
     #   print('loop stopped')
        if dead > 0:
            store_result(i, True)
        else:
            store_result(i, False)
            
def calc_ratio(i):
    #print('calc data: ',data[i])
    if data[i][1] != 0:
        data[i][4] = data[i][2]/(data[i][2]+data[i][3])
        results[data[i][0]][data[i][1]] = data[i][4]
    else:
        data[i][4] = 1
        results[data[i][0]][data[i][1]] = data[i][4]

   # print('calc data: ',data[i])
    

#print(len(data))
def run_calc():
    for i in range(len(data)):  
        if (i/(size*size))%0.01:
            print('Percentage Complete: ', i/(size*size))
        for j in range(repititions):
            battle(i)  
        calc_ratio(i)

        
run_calc()
print(data)
print(results)
```
